# Tic Tac Toe Game:

```python
import random
```

```python
#Initializing the main baord and an empty list to shoow how many moves are used.
board={1:' ',2:' ',3:' ',
       4:' ',5:' ',6:' ',
       7:' ',8:' ',9:' '}
usedmoves=[]
turnchoose=['O','X']

```

```python
#This function takes the board as input and checks if either 'O' or 'X has won the game.
def win(b):
    win=False 
    if(b[1]==b[2]==b[3] and (b[1] or b[2] or b[3])!=' ') :
        win=True
        return win
    if(b[4]==b[5]==b[6] and (b[4] or b[5] or b[6])!=' '):
        win=True
        return win
    if(b[7]==b[8]==b[9] and (b[7] or b[8] or b[9])!=' '):
        win=True
        return win
    if(b[1]==b[4]==b[7] and (b[1] or b[4] or b[7])!=' '):
        win=True
        return win
    if(b[2]==b[5]==b[8] and (b[2] or b[5] or b[8])!=' '):
        win=True
        return win
    if(b[3]==b[6]==b[9] and (b[3] or b[6] or b[9])!=' '):
        win=True
        return win
    if(b[1]==b[5]==b[9] and (b[1] or b[5] or b[9])!=' '):
        win=True
        return win
    if(b[3]==b[5]==b[7] and (b[3] or b[5] or b[7])!=' '):
        win=True
        return win
    
    

    
```

```python
#Defines the rules of the game and how to play the game.
def rules():
    print("Rules to play the game:\nThe computer will show which user (O or X) goes first and will ask you where to place your sign.")
    print("Next enter the number where you want to place your sign, if you enter a number which is already taken, the game will prompt you to enter again.")
    print(" ")
    print("Numbering used to play the game.\nEach number corresponds to its position on the board.")
    print(" ")
    print('1' + '|' +'2' + '|' + '3')
    print('-+-+-')
    print('4' + '|' + '5' + '|' + '6')
    print('-+-+-')
    print('7' + '|' + '8' + '|' + '9')
    print("           ")
    print("           ")
```

```python
# Acts as a datastructure.
#Shows the current state of the Tic Tac Toe board.
def displayboard(b): 
    print(" ")
    print("Tic Tac Toe Board:")
    print(" ")
    print(b[1] + '|' + b[2] + '|' + b[3])
    print('-+-+-')
    print(b[4] + '|' + b[5] + '|' + b[6])
    print('-+-+-')
    print(b[7] + '|' + b[8] + '|' + b[9],end='')
    print("        ")  
    print("        ") 
    print("        ") 
```

```python
#Main Program

#Choosing random sign (O or X).
turn=turnchoose[random.randint(0,1)]

#Display Rules of the game.
rules()

#Loop will iterate 9 times as there are total 9 moves possible.
for i in range(9):
    print(displayboard(board))
    
    while(True):
        move=int(input('Turn for '+turn+'. Where do you want to move?'))
        if(move not in usedmoves):
            usedmoves.append(move)
            board[move]=turn
            break
        else:
            print("That slot is already taken. Please choose another number.")
            continue
            
            
#Check is any player has won the game or not.
    result=win(board)
    if(result==True):
        print(displayboard(board))
        print("**************************************************")
        print("***************Game is WONNNN by",turn,'***************')
        print("**************************************************")

        
#Reset the board to blank.
        board={1:' ',2:' ',3:' ',
               4:' ',5:' ',6:' ',
               7:' ',8:' ',9:' '}
    
#Reset Used moves to empty.
        usedmoves=[]
    
#Break the loop and end the whole game.
        break
        
#Exchnaging the sign between O and X.    
    if(turn=='X'):
        turn="O"
    else:
        turn='X'




```
